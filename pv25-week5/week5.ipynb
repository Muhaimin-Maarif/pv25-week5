```python
import sys
import re
from PyQt6.QtWidgets import (QApplication, QWidget, QFormLayout, QLineEdit, QTextEdit, QComboBox, QPushButton,
                             QMessageBox, QLabel
                             )

class Task5(QWidget):
    def __init__(self):
        super().__init__()
        self.resize(400,400)
        self.setWindowTitle("Profile")

        self.name = QLineEdit()
        self.email = QLineEdit()
        self.age = QLineEdit()
        self.phone = QLineEdit()
        self.phone.setText("+62 ")
        self.address = QTextEdit()
        self.gender = QComboBox()
        self.gender.addItems(["","Male","Female"])
        self.education = QComboBox()
        self.education.addItems(["","SMP","SMA","S1"])
        self.save = QPushButton("Save")

        form = QFormLayout()
        form.addRow("Name: ", self.name)
        form.addRow("Email: ", self.email)
        form.addRow("Age: ", self.age)
        form.addRow("Phone Number: ", self.phone)
        form.addRow("Address: ", self.address)
        form.addRow("Gender: ", self.gender)
        form.addRow("Education: ", self.education)
        form.addRow(self.save)
        form.addRow(QLabel("Created By Muhaimin"))
        self.save.clicked.connect(self.validate)

        self.setLayout(form)

    def validate(self):
        if not self.name.text():
            self.tampil("Name is required")
            return
        
        email = self.email.text()
        if not email:
            self.tampil("email is required"); return
        elif "@" not in email or ".com" not in email:
            self.tampil("format salah"); return
        
        age = self.age.text()
        if not age:
            self.tampil("age is required"); return
        elif not age.isdigit():
            self.tampil("Harus angka semua"); return
        
        phone = self.phone.text()
        if not phone:
            self.tampil("Phone is required"); return
        elif not re.match(r'^\+?\d[\d\s\-]*$', phone):
            self.tampil("phone format is invalid"); return
        elif len(re.sub(r'\D', '', phone)) != 11:
            self.tampil("Phone number must be exactly 12 digits."); return

        if not self.address.toPlainText():
            self.tampil("address is required"); return
        
        if not self.gender.currentText():
            self.tampil("gender is required"); return
        
        if not self.education.currentText():
            self.tampil("education is required"); return

        
    def tampil(self, message):
        QMessageBox.warning(self, "Validate Error", message)

if __name__=="__main__":
    app = QApplication(sys.argv)
    window = Task5()
    window.show()
    sys.exit(app.exec())
```
